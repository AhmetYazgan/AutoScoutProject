*Second way to filling by a def function

```python
def extract_make(make_model):
    match = re.search(r'^(.*?)\s', make_model)
    return match.group(1) if match else None
```

```python
df1['Model'] = df.apply(lambda row: extract_make(row['Make_Model']) if pd.isnull(row['Model']) else row['Model'], axis=1)
```

```python
def fill_missing_column_1(df, column):
    mode_column = df.groupby('Model')[column].agg(lambda x: pd.Series.mode(x).values[0])
    model_column_map = mode_column.to_dict()
    
    return model_column_map
```

```python
model_gearbox_map = fill_missing_column_1(df1,'Gearbox')
model_gearbox_map
```

```python
df[df.Model=='A180']['Gearbox'].mode()
```

* In above, I find all mode values of Gearbox for each Model
* In below, I assigned that mode values to each missing values

```python
df['Gearbox'] = df.apply(lambda row: model_gearbox_map[row['Model']] if pd.isnull(row['Gearbox']) else row['Gearbox'], axis=1)
```
